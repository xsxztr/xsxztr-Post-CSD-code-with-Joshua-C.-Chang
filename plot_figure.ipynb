# Figure Plot

```julia
using CSV
using Plots
gr()
using DataFrames
```

# mit=0.25 L=0 case

```julia
df = CSV.read("benchmark.csv")
p1 = plot(df[:t]/60,df[:Ca_cyt],label="Ca_cyt")
p3 = plot(df[:t]/60,df[:other_Ca_cyt_tot],label="Ca_cyt_tot")

p2 = plot(df[:t]/60,df[:Ca_er],label="Ca_er")
p4 = plot(df[:t]/60,df[:other_Ca_er_tot],label="Ca_er_tot")

plot(p1,p2,p3,p4)
```

```julia
pyplot()
plt1=plot(df[:t]/60,df[:Ca_cyt],label="Ca_cyt")
xlabel!("Time (mins)")
ylabel!("Ca_cyt (muM)")
#display(plt1)
savefig(plt1, "Ca_cyt.eps")
```

# Faster NCX case MIT=0.2 L =0.5

```julia
df2 = CSV.read("fasterNCX.csv")
p1 = plot(df2[:t]/60,df2[:Ca_cyt],label="Ca_cyt")
p3 = plot(df2[:t]/60,df2[:other_Ca_cyt_tot],label="Ca_cyt_tot")

p2 = plot(df2[:t]/60,df2[:Ca_er],label="Ca_er")
p4 = plot(df2[:t]/60,df2[:other_Ca_er_tot],label="Ca_er_tot")

plot(p1,p2,p3,p4)
```

```julia
pyplot()
plt1=plot(df2[:t]/60,df2[:Ca_cyt],label="Ca_cyt")
xlabel!("Time (mins)")
ylabel!("Ca_cyt (muM)")
#display(plt1)
savefig(plt1, "Ca_cyt_fastNCX.eps")
```
